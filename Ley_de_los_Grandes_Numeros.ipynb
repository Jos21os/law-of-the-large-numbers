#The law of large numbers simply explained

In this exercise I explain in a simple way the law of large numbers with the throwing of dice and their probabilities of showing the 6 sides of the die.

## We will import Matplotlib and Empiricaldist

```python
import matplotlib.pyplot as plt
import empiricaldist
```

## We assign the dice variable with empiricaldist  

```python
dice = empiricaldist.Pmf.from_seq([1, 2, 3, 4, 5, 6])
dice.bar()
```

## Apply "for" to grow the samples

```python
for sample_size in (1e2, 1e3, 1e4):
  sample_size = int (sample_size)
  values = dice.sample(sample_size)
  print(values)

```

## We use the matplotlib library to plot the samples of the previous step

```python
for sample_size in (1e2, 1e3, 1e4):
  sample_size = int (sample_size)
  values = dice.sample(sample_size)
  sample_pmf = empiricaldist.Pmf.from_seq(values)

  plt.figure(figsize=(5, 5))
  sample_pmf.bar()
  plt.axhline(y= 1/6, color = 'red', linestyle = 'dashed')
  plt.ylim([0, 0.50])
  plt.title(f"sample_size: {sample_size}")
```
